```python
import requests
from bs4 import BeautifulSoup

# Set the search query
search_query = 'Introduction to IoT' 

# Set the base URL and search endpoint
base_url = 'https://www.cambridge.org'
search_endpoint = '/core/publications/books'

# Perform a search request
search_url = base_url + search_endpoint
response = requests.get(search_url)
soup = BeautifulSoup(response.text, 'html.parser')

# Find the search input field and get the searchWithinIds value
search_input = soup.select_one('input[name="q"]')
search_within_ids = soup.select_one('input[name="searchWithinIds"]')['value']
print(search_within_ids)

search_params= '/listing?q='+search_query.replace(' ', '+')+'&searchWithinIds='+search_within_ids

# Perform the search request with the search parameters
# status, response = http.request('http://www.nytimes.com')
response = requests.get(search_url+ search_params)
soup = BeautifulSoup(response.text, 'html.parser')

# Get the book link from the search results
print(soup.title)
# book_link = soup.select_one('.representation overview search')
getBody=soup.body

bookPageLink=[]
for link in soup.findAll("a", {"part-link"}):
    bookPageLink.append(link['href'])
    
#We are getting the page link /core/books/scientific-writing-and-publishing/92765D918D99015011DD71071C8CFD90
print(bookPageLink[0])

bookPage = requests.get(base_url+bookPageLink[0])
soup = BeautifulSoup(bookPage.text, 'html.parser')

price= soup.find("ul", {"button-group"})

# print(price.find('span').text)
print("Price: "+price.find("div", {"margin-bottom"}).text)

details= soup.find("ul", {"spec"})
print(details.text)

Author= soup.find("span", {"author-name"})
print("Author: "+Author.text)

# print("Details: "+soup.find("div", {"details"}).text)
```
